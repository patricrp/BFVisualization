```python
import pandas as pd
```

# Data Exploration for Black Friday dataset 
Celebration - November 27th

```python
df_bf = pd.read_csv('output/BlackFriday.csv')
```

```python
df_bf.head()
```

```python
# How many records are there?
df_bf.shape
```

```python
# How many duplicated are there?
df_bf.duplicated().sum()
```

```python
#Looking for null values
df_bf.isnull().sum()
```

```python
#Reducing the dataframe to only key columns

df_bf = df_bf[['query', 'id', 'date', 'user', 'followers', 'text', 'hashtags', 'url', 'urls', 'favorites', 'retweets', 'location']]
```

```python
df_bf.head()
```

```python
#Order the dataframe by date from oldest to newest tweet
df_bf.sort_values('date', ascending=True).head()
```

# Key info from the dataframe

```python
#The most engaging tweet by retweets
df_bf[(df_bf['retweets'] == df_bf["retweets"].max())]
```

```python
#Top 5 tweets by retweets
df_bf.nlargest(5, 'retweets')
```

```python
#The most influencer by number of followers
df_bf[(df_bf['followers'] == df_bf["followers"].max())]
```
